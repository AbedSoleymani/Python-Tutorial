**Try/Except** is for catching errors in python. These blocks are used for exception handling. They allow you to catch and handle exceptions that might occur during the execution of your code, rather than simply crashing your program.

Here are some benefits of using try/except in Python:

1.   **Graceful handling of errors:** By using try/except, you can handle errors in a way that doesn't interrupt the flow of your program. Instead of crashing with a traceback, you can catch the exception, log it, and continue executing the rest of the program.
2.   **Better debugging:** By catching and logging exceptions, you can get more detailed information about what went wrong in your code. This can make it easier to debug problems and fix them. *This also make the core more understandable when you come back to it after few months or when you pass it to another person in anpther department.*
3.   **Robustness:** Exception handling with try/except can make your program more robust by preventing crashes due to unforeseen circumstances. For example, if your program depends on user input or external data, it can't always be guaranteed that the input or data will be valid. Exception handling can help you handle these cases gracefully.
4.   **Flexibility:** try/except blocks allow you to handle specific exceptions separately, allowing you to tailor your error handling to the specific needs of your program.

In summary, using try/except blocks in Python can make your code more resilient, easier to debug, and more flexible.



```python
try:
  number1 = int(input('Please input the first number: '))
  number2 = int(input('Please input the second number (not zero): '))
  print('devision result is: ', number1/number2)
except ValueError:
  print('Invalid input!')
except ZeroDivisionError:
  print('Devision by zero! Please do not input 0 for the second number.')
# catch-all except block that will handle any other type of error (not aforementioned customized errors) that might occur during the code execution.
except Exception as e: 
  print("An error occurred:", e)
```

**Classes:** Classes are a fundamental feature of object-oriented programming in Python and offer several benefits, including:

1. Encapsulation: Classes allow you to group data and functions that operate on that data into a single unit. This helps you to organize your code and make it easier to understand and maintain.
2. Abstraction: By defining a class, you can abstract away complex details of the implementation and provide a simpler interface for users of the class to interact with. This allows you to create more modular and reusable code.
3. Inheritance: Classes can be inherited from each other, allowing you to create new classes that inherit the properties and methods of existing classes. This can save time and effort when creating new code, and can help to ensure consistency across related classes.
4. Polymorphism: With polymorphism, you can write code that can work with objects of different classes in a uniform way. This can simplify your code and make it more flexible.
5. Data Hiding: Classes can use access modifiers like public, private, and protected to control access to their data and methods. This helps to ensure that the data is used in a safe and consistent way and helps to prevent accidental modification or misuse of data.

Overall, classes provide a powerful and flexible way to organize and structure your code, making it more reusable, modular, and easier to maintain.

In Python classes, `self` is a special parameter that refers to the instance of the class. When you create an object from a class, Python automatically passes the object as the first argument to the method call. By convention, this parameter is usually called `self`.

The `self` parameter is used within a class to refer to the instance of the class that is being acted upon. It allows you to access the instance's properties and methods.

**A Good Example:** Building a multiple-choice quiz

```python
class Quiz:
  def __init__(self, prompt, answer):
    self.prompt = prompt
    self.answer = answer

def quiz_run(questions_answers):
  score =  0
  for question_answer in questions_answers:
    answer = input(question_answer.prompt)
    if answer == question_answer.answer:
      score += 1
  final_score = score/len(questions_answers) * 100
  return final_score
```

```python
question_prompts = [
    'What color are apples?\n(a) red\n(b) yellow\n(c) green\n',
    'What color are bananas?\n(a) red\n(b) yellow\n(c) green\n',
    'What color are cucumbers?\n(a) red\n(b) yellow\n(c) green\n'
]

questions_answers = [
    Quiz(question_prompts[0], 'a'),
    Quiz(question_prompts[1], 'b'),
    Quiz(question_prompts[2], 'c')
]

final_score = quiz_run(questions_answers)
print('Your final score is: ', str(final_score), 'out of 100.')
```

**Inheritance:** Inheritance is a concept in object-oriented programming (OOP) that allows a class to inherit properties and methods from another class. In Python, inheritance is implemented using the keyword class followed by the name of the new class, and then the name of the base class in parentheses. 

The benefits of inheritance in Python include:

1. Code reuse: Inheritance allows you to reuse code from existing classes. Instead of writing the same code multiple times in different classes, you can define a base class with the shared functionality and have other classes inherit from it.
2. Polymorphism: Inheritance allows you to create objects that can take on multiple forms. For example, you can have a Bird class and a Duck class that both inherit from an Animal class. They can each have their own unique methods and properties, but they can also be treated as Animal objects.
3. Simplification: Inheritance can simplify the code by grouping similar methods and properties into a single class, making it easier to manage and maintain.

Overall, inheritance is a powerful tool in Python that allows you to create more efficient, maintainable, and flexible code.

```python
"""
Q: Write a library of shapes in python including square, rectangle, and circle
with their attributes and functions for calculating their areas.
"""
import math

class Shapes:
  def __init__(self):
    pass
    
  def area(self):
    pass

class Rectangle(Shapes):
  def __init__(self, width, length):
    self.width = width
    self.length = length

  # Over-riding the function 'area' from Shapes class.
  def area(self):
    return self.width * self.length

class Square(Rectangle):
  def __init__(self, side):
    super().__init__(side, side)


class Circle(Shapes):
    def __init__(self, radius):
        self.radius = radius
    
    # Over-riding the function 'area' from Shapes class.
    def area(self):
        return math.pi * (self.radius ** 2)
```

```python
circle = Circle(1)
square = Square(7)
rectangle = Rectangle(2,3)

print('circle area: ', circle.area())
print('square area: ', square.area())
print('rectangle area: ', rectangle.area())
```

Another interesting example for inheritance:

```python
class Person:
  def __init__(self, name, age, gender):
    self.name = name
    self.age = age
    self.gender = gender
  
  def introduce(self):
    print(f"Hi, my name is {self.name}, and I'm a {self.age} year old {self.gender}.")

class Student(Person):
  def __init__(self, name, age, gender, grade):
    super().__init__(name, age, gender)
    self.grade = grade
  
  def introduce(self):
    super().introduce()
    print(f"I'm a student in grade {self.grade}.\n")

class Teacher(Person):
  def __init__(self, name, age, gender, subject):
    super().__init__(name, age, gender)
    self.subject = subject
  
  def introduce(self):
    super().introduce()
    print(f"I'm a {self.subject} teacher.\n")
```

```python
s = Student("Alice", 16, "female", 11)
t = Teacher("Bob", 35, "male", "math")

s.introduce() 
t.introduce()
```
